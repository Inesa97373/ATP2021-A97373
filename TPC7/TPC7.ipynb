### Resumo:
Este trabalho, proposto na UC de Algoritmos e Técnicas de Programação foi realizado na linguagem Python. Este consiste num conjunto de funções que fazem a leitura de datasets e do seu processamento nomeadamente consulta de dataset, top10 de médias, distribuição por curso, distribuição por médias; e a visualização de gráficos das distribuições dos cursos e médias.


### Código:

```python
#BDalunos=[Aluno]
#Aluno={id: , nome:  , curso:   ,tpc:[int]}

# Leitura/carregamento da informação do ficheiro

def lerDataset(fnome):
    bd=[]
    f=open(fnome, encoding='utf-8')
    f.readline()
    for linha in f:
        
        novaLinha = linha.replace('\n', '')
        campos = novaLinha.split(",")
        aluno = {}
        aluno['id'] = campos[0]
        aluno['nome'] = campos[1]
        aluno['curso'] = campos[2]
        listaTPC=[]
        for tpc in campos[3:]:
            listaTPC.append(int(tpc))
        aluno['tpc']=listaTPC
        bd.append(aluno)
    
    return bd

BDalunos=lerDataset("alunos.csv")
print(BDalunos)

# Listagem da informação

def listarDataset(bd):
    print("id        |nome            |curso       |média")
    print("--------------------------------------------------------")
    for a in bd:
        print(a['id'] + '|' + a['nome'] + '|' + a['curso'] + '|' + str((sum(a['tpc'])/len(a['tpc']))))
              
listarDataset(BDalunos)


# Consulta de um registo
def consultarDataset(bd, id):
    for a in bd:
        if a['id']==id:
            print(a['id'] + '  |  ' +  a['nome']+ '   |   ' + a['curso']  + '   |   ' + str(a['tpc']) + '   |  ' + str((sum(a['tpc']))/len(a['tpc'])))
              
            
consultarDataset(BDalunos, '"a1"')


def top10(bd):
    novaBD=[]
    for a in bd:
        novoAluno= a
        novoAluno['media']= (sum(a['tpc']))/len(a['tpc'])
        novaBD.append(novoAluno)
    listaOrdenada= sorted(novaBD, key=lambda d: d['media'], reverse=True)
    
    i=0
    while i<10:
        a=listaOrdenada[i]
        print(a['id'] + '  |  ' +  a['nome']+ '   |   ' + a['curso']  + '   |  ' + str(a['media']))
        i+=1
        
        
top10(BDalunos)  


def distribPorCurso(bd):
    distribuicao = {}
    for a in bd:
        if a['curso'] in distribuicao.keys():
            distribuicao[a['curso']]=distribuicao[a['curso']]+1
        else:
            distribuicao[a['curso']]=1
            
    return distribuicao
distribPorCurso(BDalunos)



def distribPorMedia(bd):
    distribuicao = {}
    for a in bd:
        media= int((sum(a["tpc"])/len(a["tpc"])))
        if media in distribuicao.keys():
            distribuicao[media]=distribuicao[media]+1
        else:
            distribuicao[media]=1
            
    return distribuicao
distribPorMedia(BDalunos)



import matplotlib.pyplot as plt

def plotDistribPorCurso(bd):
    distribuicao=distribPorCurso(bd)
 
    x = list(range(1,len(distribuicao.keys())+1))
    y = list(distribuicao.values())

    cursos = list(distribuicao.keys())

    plt.bar(x, y, tick_label = cursos, width = 0.8, color = ['aquamarine', 'violet', 'pink', 'beige'])

    plt.xlabel('Cursos')
    plt.ylabel('Nº alunos')
    plt.title('Distribuição por cursos')

    plt.show()

plotDistribPorCurso(BDalunos)


def plotDistribPorMedia(bd):
    distribuicao=distribPorMedia(bd)
 
    x = list(range(1,len(distribuicao.keys())+1))
    y = list(distribuicao.values())
    medias = list(distribuicao.keys())

    plt.bar(x, y, tick_label = medias, width = 0.8, color=['lightcoral', 'skyblue', 'khaki'])

    plt.xlabel('Médias')
    plt.ylabel('Nº alunos')
    plt.title('Distribuição por média')
   
    plt.show()

plotDistribPorMedia(BDalunos)
```
