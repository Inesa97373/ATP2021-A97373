### Resumo
Este trabalho, proposto na UC de Algoritmos e Técnicas de Programação foi realizado na linguagem Python e tem como objetivo criar um algoritmo capaz de trabalhar com funções e operações básicas entre frações (soma, subtração, multiplicação e divisão), e ainda com listas, tendo a opção de criar uma lista de frações aleatória ou criar uma lista manualmente.
O programa apresenta um menu com 8 opções, em que é pedido ao utilizador para escolher a que pretende. O algoritmo apresenta 8 funções:
* __criarFracao:__ permite criar uma fração
* __mdc:__ calcula o máximo divisor comum de dois números, que irá ser usada para simplificar as frações
* __verFracaoSimplificada:__ faz print da fração criada na sua forma simplificada 
* __somar:__ faz a soma da fração criada com a nova fração que o utilizador vai criar
* __subtrair:__ faz a subtração da fração criada com a nova fração que o utilizador vai criar
* __multiplicar:__ faz a multiplicação da fração criada com a nova fração que o utilizador vai criar
* __dividir:__faz a divisão da fração criada com a nova fração que o utilizador vai criar 
* __lista:__permite criar uma lista de frações de forma aleatória ou de forma manual e apresenta essas frações na forma simplificada


### Código:

```python
import random
def menu():
    print("""
          (1) criar uma fração
          (2) ver fração simplificada
          (3) somar frações
          (4) subtrair frações
          (5) multiplicar frações
          (6) dividir frações
          (7) criar uma lista de frações
          (0) sair""")
    
def criarFracao():
    n=int(input("Introduza o numerador: "))
    d=int(input("Introduza o denominador: "))
    f=(n,d)
    return(f)

def verFracaoSimplificada(f):
    n,d=f
    m=mdc(n,d)
    n=int(n/m)
    d=int(d/m)
    return(str(n)+"/"+str(d))

def mdc(a,b):
    while(b!=0):
        resto=a%b
        a=b
        b=resto
    return a

def somar(f1,f2):
    n1,d1=f1
    n2,d2=f2
    f=n1*d2+n2*d1, d1*d2
    return verFracaoSimplificada(f)

def subtrair(f1,f2):
    n1,d1=f1
    n2,d2=f2
    f=n1*d2-n2*d1, d1*d2
    return verFracaoSimplificada(f)

def multiplicar(f1,f2):
    n1,d1=f1
    n2,d2=f2
    f=n1*n2, d1*d2
    return verFracaoSimplificada(f)

def dividir(f1,f2):
    n1,d1=f1
    n2,d2=f2
    f=n1*d2, d1*n2
    return verFracaoSimplificada(f)

def lista():
    lis=[]
    l=input("Pretende criar uma lista aleatória de frações (a) ou manualmente (m)? ")
    if l=="a":
        tam=int(input("Quantos elementos pretende que a lista contenha? "))
        for i in range(tam):
            num = random.randint (1,100)
            den = random.randint (1,100)
            f=num,den
            lis.append(verFracaoSimplificada(f))

    if l=="m":
        tam=int(input("Quantos elementos pretende que a lista contenha? "))
        for i in range(tam):
             num=int(input("Introduza o numerador: "))
             den=int(input("Introduza o denominador: "))
             f=num,den
             lis.append(verFracaoSimplificada(f))
    return lis
        
op="1"
menu()

while (op!= "0"):
    op= input ("O que pretende fazer? ")
    if op=="1":    
        f1=criarFracao()

    elif op=="2":
        print(verFracaoSimplificada(f1))
        
    elif op=="3":
        print("Crie a fração que quer somar com a primeira criada. ")
        f2=criarFracao()
        soma=somar(f1,f2)
        print(soma)
        
    elif op=="4":
        print("Crie a fração que quer subtrair à primeira criada. ")
        f2=criarFracao()
        subtracao=subtrair(f1,f2)
        print(subtracao)
        
        
    elif op=="5":
        print("Crie a fração que quer multiplicar com a primeira criada. ")
        f2=criarFracao()
        multiplicacao=multiplicar(f1,f2)
        print(multiplicacao)
         
        
    elif op=="6":
        print("Crie a fração que quer dividir pela primeira criada. ")
        f2=criarFracao()
        divisao=dividir(f1,f2)
        print(divisao)
        
    elif op=="7":
        lis1=lista()
        print (lis1)
         
       
    elif op=="0":
        print ("O programa vai encerrar. Até à próxima! ")
    else:
        print ("Opção inválida! ")
```
